```python
import os, sys
import numpy as np
from keras_xlnet import PretrainedList, get_pretrained_paths
from keras_xlnet import Tokenizer

sys.path.insert(0, "src")

from xlnet.dataloader import load_data, get_X_array
from xlnet.model import get_xlnet_model
```

```python
os.environ['KMP_DUPLICATE_LIB_OK']='True'

BATCH_SIZE = 20
SEQ_LEN = 128
LR = 5e-6
```

```python
paths = get_pretrained_paths(PretrainedList.en_cased_base)
tokenizer = Tokenizer(paths.vocab)

model = get_xlnet_model(paths, BATCH_SIZE, SEQ_LEN, LR)
model.load_weights("result/xlnet_voted_up.best.h5")
```

```python
from xlnet.model_eval import model_eval, plot_roc
```

```python
X, y = load_data(tokenizer, 'data/reviews_112_trans-en.jl', SEQ_LEN=SEQ_LEN, target_label='voted_up')
```

```python
preds = model.predict(get_X_array(X), verbose=True, batch_size=BATCH_SIZE)
```

```python
model_eval(preds.argmax(axis=-1), y)
```
